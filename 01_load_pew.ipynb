```python
import pandas as pd
import numpy as np

# 1) Human-readable labels for columns (question text)
col_labels = dict(zip(meta.column_names, meta.column_labels))

# 2) Value-label dictionaries (SPSS codes -> text per column)
# meta.variable_value_labels is a dict: {colname: {code: "label", ...}, ...}
val_labels = meta.variable_value_labels

def labelize(s: pd.Series, colname: str):
    """Map SPSS-coded values to their text labels when available."""
    mapping = val_labels.get(colname, {})
    # Convert keys to same dtype as the series’ values
    mapping = { (type(s.dropna().iloc[0])(k) if len(s.dropna()) else k): v for k, v in mapping.items() }
    return s.map(mapping).fillna(s)

def tab(s: pd.Series, colname: str, normalize=False, dropna=True, top=20):
    """Quick frequency table, labeled if possible."""
    x = s.copy()
    if dropna:
        x = x[~x.isna()]
    out = x.value_counts(normalize=normalize).head(top)
    out.index = labelize(out.index.to_series(), colname)
    return out
```

```python
search_terms_ai = ["ai", "artificial", "automation", "robot", "technology", "privacy", "algorithm", "jobs"]
search_terms_demo = ["age", "educ", "education", "income", "party", "gender", "race", "hispanic"]

ai_vars = [(k, v) for k, v in col_labels.items()
           if any(t in str(v).lower() for t in search_terms_ai)]
demo_vars = [(k, v) for k, v in col_labels.items()
             if any(t in (str(k).lower() + " " + str(v).lower()) for t in search_terms_demo)]

print("AI/Tech-related variables (first 15):")
for k, v in ai_vars[:15]:
    print(f"- {k}: {v}")

print("\nDemographics candidates (first 15):")
for k, v in demo_vars[:15]:
    print(f"- {k}: {v}")
```

```python
# Map coded values -> text labels for readability
df["ai_exc_concern"] = df["CNCEXC_W99"].map(meta.variable_value_labels.get("CNCEXC_W99", {}))
df["age_lbl"]  = df["AGE"].map(meta.variable_value_labels.get("AGE", {}))
df["educ_lbl"] = df["EDUC"].map(meta.variable_value_labels.get("EDUC", {}))

# Quick look at distributions
print("AI Excited vs Concerned:")
display(df["ai_exc_concern"].value_counts(dropna=False))

print("\nAge groups:")
display(df["age_lbl"].value_counts(dropna=False).head(10))

print("\nEducation levels:")
display(df["educ_lbl"].value_counts(dropna=False).head(10))
```

```python
# look for likely demographic columns by NAME or by LABEL text
def find_candidates(needles):
    needles = [n.lower() for n in needles]
    hits = []
    for k, v in col_labels.items():
        name = str(k).lower()
        lab  = str(v).lower()
        if any(n in name for n in needles) or any(n in lab for n in needles):
            hits.append((k, v))
    return hits

age_cands  = find_candidates(["age"])
educ_cands = find_candidates(["educ","education"])
print("AGE candidates:")
for k,v in age_cands:  print("-", k, "→", v)
print("\nEDUC candidates:")
for k,v in educ_cands: print("-", k, "→", v)
```

```python
df["ai_exc_concern"] = df["CNCEXC_W99"].map(meta.variable_value_labels.get("CNCEXC_W99", {}))
df["age_lbl"]  = df["F_AGECAT"].map(meta.variable_value_labels.get("F_AGECAT", {}))
df["educ_lbl"] = df["F_EDUCCAT"].map(meta.variable_value_labels.get("F_EDUCCAT", {}))
```

```python
ct_age  = pd.crosstab(df["age_lbl"],  df["ai_exc_concern"], normalize="index") * 100
ct_educ = pd.crosstab(df["educ_lbl"], df["ai_exc_concern"], normalize="index") * 100

print("AI attitude by AGE (%)")
display(ct_age.round(1))

print("\nAI attitude by EDUC (%)")
display(ct_educ.round(1))
```
